```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
# pip install selenium
from selenium import webdriver
from selenium.webdriver.chrome.service import Service

service = Service("/usr/local/bin/chromedriver")
drive = webdriver.Chrome(service=service) 
drive.get('https://www.immoweb.be/en/classified/apartment/for-sale/woluwe-saint-lambert/1200/11092028')
soup = BeautifulSoup(drive.page_source, "lxml")
jobs_elements = soup.find_all("tr", class_= "classified-table__row")
```

```python

#start to scrape
data = []

for jobs_element in jobs_elements:
    try:
        table_data = jobs_element.find("td", class_="classified-table__data").text.strip()
    
    # if you dont add except .text method will produce an error
    except AttributeError:
        table_data = 'no data'
    
    try:
        table_header = jobs_element.find("th", class_="classified-table__header").text.strip()
    
    # if you dont add except .text method will produce an error
    except AttributeError:
        table_header = 'no data'
    
    item = {'table_data':table_data,
            'table_header': table_header}
    
    data.append(item)
    
print(data)

df = pd.DataFrame(data)

df
```
