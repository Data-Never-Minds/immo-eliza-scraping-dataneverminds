```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from matplotlib.ticker import FuncFormatter

# Supondo que 'df' seja seu DataFrame
# Carregue seus dados em um DataFrame conforme necessário
# df = pd.read_csv('path_to_your_data.csv') # Para um arquivo CSV
# df = pd.read_excel('path_to_your_data.xlsx') # Para um arquivo Excel
# df = pd.read_sql(query, connection) # Para dados de um banco de dados SQL

# Criando o histograma com uma sobreposição KDE
plt.figure(figsize=(10, 6))
sns.histplot(data['Price'], bins=50, kde=True)

# Definindo o título e os rótulos do gráfico
plt.title('Distribution of Property Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.xlim(0, 2500000)

# Função para formatar os rótulos de preço
def price_formatter(x, pos):
    return 'R${:,.0f}'.format(x)

# Aplicando a formatação dos rótulos de preço ao eixo x
plt.gca().xaxis.set_major_formatter(FuncFormatter(price_formatter))

# Mostrando o gráfico5
plt.show()


data['Teste'] = 'teste'

data
```
